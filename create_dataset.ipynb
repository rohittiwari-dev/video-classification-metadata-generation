```python

import os
import pandas as pd
from glob import glob 
from IPython.display import clear_output
import cv2
import shutil
```

```python
exceldf=pd.read_excel("/content/drive/MyDrive/Isro_project/Video-Classification.xlsx")
```

```python
video_path=glob("/content/drive/MyDrive/Isro_project/Source Video/*.mp4")
save_path="/content/drive/MyDrive/Isro_project/Video Set"


if video_path:
    listfiles=os.listdir("/content/drive/MyDrive/Isro_project/Source Video")
    file_count=len(listfiles)
    process='.'
    ratio=0
    for path in video_path:
        video_file_name=os.path.split(path)[-1]
        process+=process
        ratio+=1
        for ind,row in exceldf.iterrows():
            
            if row['Filename']==video_file_name:
                category=row['Category']
                save_path1=os.path.join(save_path,category)
                if not os.path.exists(save_path1):
                    os.makedirs(save_path1)
                shutil.copy(path,os.path.join(save_path1,video_file_name))
            
        print("************************************************")
        print(f"Creating Video DataSet{process}")
        print(f"Process Completion:{round((ratio/file_count)*100,2)} %")
        print("***************************************************")
        clear_output(wait=1)
        if len(process)>4:
            process='.'
```

```python
src_path="/content/drive/MyDrive/Isro_project/Video Set"
save_path="/content/drive/MyDrive/Isro_project/Processing"

for root,dirs,files in os.walk(src_path):
    process='.'
    ratio1=0
    for category in dirs:
        listfiles=os.listdir(src_path)
        file_count=len(listfiles)

        saved_path=f"{src_path}/{category}"
        listvid=os.listdir(saved_path)
        vid_count=len(listvid)
        video_path=glob(saved_path+"/*.mp4")
        process+=process
        if len(process)>4:
            process='.'
        ratio1+=1

        if video_path:

            ratio=0
            for video in video_path:
                ratio+=1    

                if float((ratio/vid_count)*100)<float(50):
                    saving_path=save_path+"/Training/"+category
                else:
                    saving_path=save_path+"/Validation/"+category

                video_capture=cv2.VideoCapture(video)
                fps=video_capture.get(5)
                frameid=0
                frm_id=0

                while video_capture.isOpened():
                    rtn_frame,frame =video_capture.read()
                    if not rtn_frame:
                        video_capture.release()
                        break

                    if frameid%fps==0:
                        frm_id+=1
                        name=os.path.split(video)[-1].split('.')[0]
                        name1=name+"-"+str(frm_id)
                        frame_resize=cv2.resize(frame ,(224,224))
                        if not os.path.exists(saving_path):
                            os.makedirs(saving_path)
                        colorfranme=cv2.cvtColor(frame_resize, cv2.COLOR_BGR2RGB)
                        cv2.imwrite(saving_path+"/"+name1+".jpg",colorfranme)
                        print("************************************************")
                        print(f"Creating Dataset for Traning{process}")
                        print(f"Open Folder : {category}")
                        print("Processing : ",saving_path+"/"+name1+".jpg")
                        print(f"Frame Extraction : {round((ratio/vid_count)*100,2)} %")
                        print(f"Percentage Completed : {round((ratio1/file_count)*100,2)} %")
                        print("************************************************")
                        clear_output(wait=1)
                    frameid+=1
```
